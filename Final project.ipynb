<h1 style="font-family: Times New Roman; font-size: 28px;color:DarkoliveGreen;">
Data pre-processing üßπ: </h1>

<h1 style="font-family: Times New Roman; font-size: 22px;color:DarkSlateGray;">
üîπRenaming Columns:</h1>

```python
df.columns = ["ID", "Gender", "Age", "Hypertension", "HeartDisease", 
              "Married", "WorkType", "ResidenceType", "AvgGlucoseLevel", 
              "BMI", "SmokingStatus", "Stroke"]
df
```

```python
df.rename(columns={'Stroke': 'StrokeStatus'}, inplace=True)
df
```

<h1 style="font-family: Times New Roman; font-size: 22px;color:DarkSlateGray;">
üîπReplacing:</h1>

```python
df['Gender']=df['Gender'].replace({'M':'male','F':'Female'})
df['WorkType']=df['WorkType'].replace({'Govt_job':'Government_Job'})
df
```

```python
df['StrokeStatus'] = df['StrokeStatus'].replace({1: 'Stroke occurred', 0: 'No Stroke'})
df
```

<h1 style="font-family: Times New Roman; font-size: 22px;color:DarkSlateGray;">
üîπCounting number of gender category:</h1>

```python
df['Gender'].value_counts()
```

<h1 style="font-family: Times New Roman; font-size: 22px;color:DarkSlateGray;">
üîπDroping the 'Other' Category </h>

```python
df = df[df['Gender'] != 'Other']
```

<h1 style="font-family: Times New Roman; font-size: 22px;color:DarkSlateGray;">
üîπTotal number of missing values per column:
</h1>

```python
print('Missing values:')
df.isna().sum()
```

<h1 style="font-family: Times New Roman; font-size: 22px;color:DarkSlateGray;">
üîπFilling out blank columns:
</h1>

```python
df['Age']=df['Age'].fillna(df['Age'].mean().round(2))
df['BMI']=df['BMI'].fillna(df['BMI'].median())
df['SmokingStatus']=df['SmokingStatus'].fillna('Unknown')
df['AvgGlucoseLevel'].fillna(df['AvgGlucoseLevel'].mean(), inplace=True)
df
```

<h1 style="font-family: Times New Roman; font-size: 22px;color:DarkSlateGray;">
üîπTo capitalize the first letters:</h1>

```python
df['Gender']=df['Gender'].str.capitalize()
df['SmokingStatus']=df['SmokingStatus'].str.capitalize()
df
```

<h1 style="font-family: Times New Roman; font-size: 22px;color:DarkSlateGray;">
üîπSort by ID ascending:</h1>

```python
df = df.sort_values(by='ID', ascending=True)
df
```

<h1 style="font-family: Times New Roman; font-size: 22px;color:DarkSlateGray;">
üîπTotal number of duplicate values:</h>

```python
df.duplicated().sum()
```

<h1 style="font-family: Times New Roman; font-size: 22px;color:DarkSlateGray;">
üîπTo remove the duplicate values:</h1>

```python
df.drop_duplicates(inplace=True)
df
```

<h1 style="font-family: Times New Roman; font-size: 22px;color:DarkSlateGray;">
üîπCreating new columns:</h1>

```python
#AgeGroup category
```

```python
df['AgeGroup'] = 'Unknown'

# Assigning age groups based on Age
df.loc[df['Age'] <= 18, 'AgeGroup'] = 'Child'
df.loc[(df['Age'] > 18) & (df['Age'] <= 35), 'AgeGroup'] = 'Adult'
df.loc[(df['Age'] > 35) & (df['Age'] <= 50), 'AgeGroup'] = 'Middle_Aged'
df.loc[(df['Age'] > 50) & (df['Age'] <= 65), 'AgeGroup'] = 'Senior'
df.loc[df['Age'] > 65, 'AgeGroup'] = 'Elderly'
df
```

```python
# BMI category
```

```python
df["BMICategory"] = "Unknown"

df.loc[df['BMI'] < 18.5, 'BMICategory'] = 'Underweight'
df.loc[(df['BMI'] >= 18.5) & (df['BMI'] <= 24.9), 'BMICategory'] = 'Normal'
df.loc[(df['BMI'] >= 25) & (df['BMI'] <= 29.9), 'BMICategory'] = 'Overweight'
df.loc[df['BMI'] >= 30, 'BMICategory'] = 'Obese'
df
```

```python
# Glucose Category
```

```python

df["GlucoseCategory"] = "Unknown"

df.loc[df['AvgGlucoseLevel'] < 70, 'GlucoseCategory'] = 'Low'
df.loc[(df['AvgGlucoseLevel'] >= 70) & (df['AvgGlucoseLevel'] < 100),
       'GlucoseCategory'] = 'Normal'
df.loc[(df['AvgGlucoseLevel'] >= 100) & (df['AvgGlucoseLevel'] < 126),
       'GlucoseCategory'] = 'Prediabetic'
df.loc[df['AvgGlucoseLevel'] >= 126, 'GlucoseCategory'] = 'Diabetic'
df
```
